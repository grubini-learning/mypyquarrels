```python
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.visited = [False] * (vertices + 1)
        self.history = []
        self.directions = {vertex: [] for vertex in range(vertices + 1)}

    def add_edge(self, nodea, nodeb):
        self.directions[nodea].append(nodeb)

    def get_directions(self):
        return self.directions

    def start_trip(self):
        for vertex in range(1, self.vertices + 1):
            self.reset_visited()
            #visiting each vertex   because we dont know from which we get the answer
            self.history.append(vertex)
            if self.ham_cycle(vertex):
                return [1] + self.history
        return -1

    def ham_cycle(self, city):
        #la condiciiohn par que termine en verdadero
        #la condicion para que salga y false
        current = city
        while len(self.history) > 0:
            if self.validate(current):
                if self.final_condition(current):
                    return True
                else:
                    self.visited[current] = True
                    self.history.append(current)
                    current = self.next_point(current)
            else:
                if self.final_condition(current):
                    return True
                else:
                    current = self.history.pop()
                    self.visited[current] = False
            if current == None:
                current = self.history.pop()



    def next_point(self, city):
        current = 0
        while current < len(self.directions[city]):
            neighbour = self.directions[city][current]
            if self.visited[neighbour] == False:
                return neighbour
            else:
                self.visited[neighbour] = False
                current += 1
        return None

    def final_condition(self, city):
        unvisited = self.visited.count(False)
        if unvisited == 0:
            self.history.append(city)
            return True
        else:
            return False

    def validate(self, city):
        if self.visited[city] == True and city in self.history:
            return False
        elif len(self.directions[city]) == 0:
            self.visited[city] = True
            return False
        else:
            return True

    def reset_visited(self):
        self.visited = [False if city > False else True for city in range(len(self.visited))]
        self.history.clear()
```
