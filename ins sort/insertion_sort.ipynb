```python
def ins_sort():
    arr = [6, 10, 4, 5, 1, 2]
    length = len(arr)
    swap = False
    for position in range (1, length):
        if swap:
            current = position
            previous = position - 1
        

        temp = arr[previous]
        while current < length and arr[previous] > arr[current]:
            arr[previous] = arr[current]
            previous = current
            current = current + 1
            swap = True
        
        arr[previous] = temp
    return arr
            
```

```python
def insertion_sort(arr):
    counter = 0
    current = 1
    while current < len(arr):
        previous = current - 1
        if arr[previous] > arr[current]:
            temp = arr[previous]
            while current < len(arr) and arr[previous] > arr[current]:
                counter += 1
                arr[previous] = arr[current]
                previous = current
                current = current + 1
            arr[previous] = temp
            previous = 0
            current = 1
        else:
            current = current + 1
    return counter

```
